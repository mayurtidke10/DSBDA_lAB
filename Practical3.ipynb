```python
df=pd.read_csv("crime.csv",encoding='latin1')
```

```python
df['Murder'].fillna(0,inplace=True)
```

```python
victims=df['Murder']
```

```python
desc=victims.describe()
```

```python
desc
```

```python
mean=desc['mean']
q1=desc['25%']
q3=desc['75%']
median=victims.median()
stddev=desc['std']
var=stddev**2
mode=victims.mode()
print("Mean:",mean)
print("Q1:",q1)
print("Q3:",q3)
print("Median:",median)
print("Mode:",mode)
print("Standard Deviation:",stddev)
print("Variance:",var)
```

```python
skewness=(3*(mean-median/stddev))
if(skewness>0):
    print("Right Skew")
else:
    print("Left Skew")
```
